```python
import pandas as pd
import numpy as np
from sklearn import metrics
```

```python
df=pd.read_csv('problem_2_train.csv')
```

```python
df.head(10)
```

```python
df['memcreateddate']=pd.to_datetime(df['memcreateddate'])
df['mobilepurchasedate']=pd.to_datetime(df['mobilepurchasedate'])
df['claimintdate']=pd.to_datetime(df['claimintdate'])
df['pickupStartdate']=pd.to_datetime(df['pickupStartdate'])
```

```python
df['memcreateddate']=pd.to_numeric(df['memcreateddate'])
df['mobilepurchasedate']=pd.to_numeric(df['mobilepurchasedate'])
df['claimintdate']=pd.to_numeric(df['claimintdate'])
df['pickupStartdate']=pd.to_numeric(df['pickupStartdate'])
```

```python
from sklearn.preprocessing import LabelEncoder
labelEncoder=LabelEncoder()
df['mobmake']=labelEncoder.fit_transform(df['mobmake'])
df['custpincodecategory']=labelEncoder.fit_transform(df['custpincodecategory'])
df['claimtype']=labelEncoder.fit_transform(df['claimtype'])
```

# X and y

```python
feature_cols=['hubid','pickuppartnerid','insuranceCompanyId','deliverypartnerid','claimintdate','claimtype',
      'pickupStartdate','custpincode','servicecntrid','custpincodecategory','mobmake']
```

```python
X=df[feature_cols]
```

```python
y=df['tat_in_days']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
linreg.fit(X_train,y_train)

y_pred=linreg.predict(X_test)
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

# KNN

```python
from sklearn.neighbors import KNeighborsRegressor
knnreg= KNeighborsRegressor()
knnreg.fit(X_train,y_train)

y_pred=knnreg.predict(X_test)
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

# Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
gradreg= GradientBoostingRegressor()
gradreg.fit(X_train,y_train)

y_pred=gradreg.predict(X_test)
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

# Ridge

```python
from sklearn.linear_model import Ridge
ridreg= Ridge()
ridreg.fit(X_train,y_train)

y_pred=ridreg.predict(X_test)
print(np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
